```python
from retinaface import RetinaFace
```

```python
import cv2
```

```python
resp = RetinaFace.detect_faces("img1.jpg")
resp
```

```python
for i in resp:
  print(resp[i]['facial_area'])
```

```python
img = cv2.imread("img1.jpg")
```

```python
for i in resp:
  x1,y1,x2,y2 = resp[i]['facial_area']
  cv2.rectangle(img, (x1,y1), (x2,y2), (0,255,0), thickness=2)
```

```python
cv2.imshow("img", img)
```
